## VCF results format

```python
import toytree
import ipcoal
```

### Define the species tree model

```python
tree = toytree.rtree.unittree(6, 1e6)
```

### Define the coalescent model

```python
model = ipcoal.Model(tree, Ne=1e6, nsamples=2)
```

### Simulate data

```python
model.sim_loci(10, 2000)
```

### Show results

```python
model.df
```

```python
model.write_vcf(diploid=True)
```
